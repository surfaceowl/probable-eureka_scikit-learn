```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")

# create blobs
data = make_blobs(n_samples=20000, n_features=4, centers=4, cluster_std=3.6, random_state=150)

# create np array for data points
points = data[0]

# create scatter plot
plt.figure(figsize=(18,9), dpi=120)
plt.title("Distribution of Initial Data", fontdict=None, loc='center')
plt.style.use('ggplot')
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='RdYlBu')
plt.xlabel("horizontal distance from center")
plt.xlim(-20,20)
plt.ylabel("vertical distance from center")
plt.ylim(-20,20)
```

```python
# import kmeans
from sklearn.cluster import KMeans
```

```python
# create kmeans object
kmeans = KMeans(n_clusters=4)

# fit kmeans object to data
kmeans.fit(points)

# print location of clusters learned by kmeans object
print(kmeans.cluster_centers_)

# save new clusters for chart
y_km = kmeans.fit_predict(points)
```

```python
plt.figure(figsize=(18, 9), dpi=120)
# color keys:  https://xkcd.com/color/rgb/
plt.scatter(points[y_km ==0,0], points[y_km == 0,1], s=100, c='xkcd:cadet blue')
plt.scatter(points[y_km ==1,0], points[y_km == 1,1], s=100, c='xkcd:dark sea green')
plt.scatter(points[y_km ==2,0], points[y_km == 2,1], s=100, c='xkcd:light grey blue')
plt.scatter(points[y_km ==3,0], points[y_km == 3,1], s=100, c='xkcd:butter yellow')
```
